Importing necessary libraries

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Upload Dataset

```python
Diabetes_Data = pd.read_csv('/content/diabetes.csv')
```

```python
Diabetes_Data.head()
```

```python
# number of rows and Columns in this dataset
Diabetes_Data.shape
```

```python
# Getting the statistical measures of the data
Diabetes_Data.describe()
```

```python
#Count result data info
Diabetes_Data['Outcome'].value_counts()
```

```python
#Mean value of outcome to know the statistical info
Diabetes_Data.groupby('Outcome').mean()
```

```python
#Separating the data & labels
X=Diabetes_Data.drop(columns='Outcome', axis=1)
Y=Diabetes_Data['Outcome']
```

```python
#Data Standardization (To make all the variables value range same)
scaler= StandardScaler()
```

```python
#now fit the incosistent data in the StandardScaler()
scaler.fit(X)
```

```python
#Transforming all the data in a common range
Standardized_data = scaler.transform(X)
```

```python
print(Standardized_data)
```

```python
#Putting the standardized data in to the X
X=Standardized_data
```

```python
#Tran test split(used for splitting a dataset into training and testing sets)
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.2,stratify=Y,random_state=2)
```

```python
#Check Data shape now
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Training the model
classifier= svm.SVC(kernel='linear')
```

```python
#Fit our training data to classifier
classifier.fit(X_train,Y_train)
```

```python
#Model evaluation through accuracy test
X_train_Prediction= classifier.predict(X_train)
Training_data_accuracy=accuracy_score(X_train_Prediction,Y_train)
```

```python
print('Training Accuracy: ',Training_data_accuracy)
```

```python
#Testing Accuracy
X_test_Prediction= classifier.predict(X_test)
Test_data_accuracy=accuracy_score(X_test_Prediction,Y_test)
```

```python
print('Test Accuracy: ',Test_data_accuracy)
```

```python
#Making a predicting system to predict
input_data = (5,166,72,19,175,25.8,0.587,51)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print('\n',prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('\nThe person is diabetic\n')
```
